```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# проверка У1 на соответствие нормальному распределению
print("Проверка Y1 на соответствие нормальному распределению")

stat, p = scipy.stats.normaltest(df['Y1']) # Критерий согласия Пирсона

print('Statistics=%.3f, p-value=%.3f' % (stat, p))
alpha = 0.05
stat, p = scipy.stats.normaltest(df['Y1']) # Критерий согласия Пирсона

if p>alpha:
    print("Принять гипотезу о нормальности")
else:
    print("Отклонить гипотезу о нормальности")
    
# построение графика
fig, ax = plt.subplots()
%matplotlib inline
df=df.sort_values(['Y1']) # 1) отсортировали, 2) построили гистограмчу
print(df)

# нормальное распределение и гистограмма
sns. histplot(
    df['Y1'], kde=True,
    stat="density", kde_kws=dict(cut=3),
    alpha=.4, edgecolor=(1, 1, 1, .4),
)
fig.savefig('gistY1.png', transparent=False, dpi=100, bbox_inches="tight") # сохранение рисунка в файл
plt. show()
```

```python
meanY1=df['Y1'].mean()
meanY2=df['Y2'].mean()
print(meanY1, meanY2)

stdY1=df['Y1'].std(ddof=1)
stdY2=df['Y2'].std(ddof=1)
print(stdY1,stdY2)

z=2

print('Y1')
uppY1=meanY1+z*stdY1
print('Верх граница=',uppY1)
lowY1=meanY1-z*stdY1
print('Ниж граница=',lowY1)

print('Y2')
uppY2=meanY2+z*stdY2
print('Верх граница=',uppY2)
lowY2=meanY2-z*stdY2
print('Ниж граница=',lowY2)
```

```python
df1 = df.copy()
df1['Y1'] = df1['Y1'].clip(lower=lowY1, upper=uppY1)
df1['Y2'] = df1['Y2'].clip(lower=lowY2, upper=uppY2)

sns.set(font_scale=1.15)
plt.figure(figsize=(8,4))
sns.heatmap(
    df1.corr(),        
    cmap='RdBu_r', # задаём цветовую схему
    annot=True, # рисуем значения внутри ячеек
    vmin=-1, vmax=1) # указываем начало цветовых кодов от -1 до 1.
plt.show()
```
